```python
pasajeros= df["passengers"]
media = np.mean(pasajeros)
media
```

```python
df["underperforming"]=0
```

```python
for rowname, rowserie in df.iterrows():
    if rowserie["passengers"]<media:
        df.set_value(rowname,"underperforming",1)
        
```
